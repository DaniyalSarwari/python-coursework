## Guess the Number (Computer)

- The computer randomly selects a number within a given range.
- The user guesses the number, and the computer provides feedback if the guess is too high, too low, or correct.
- The game continues until the correct number is guessed, and the total number of attempts is displayed.

## Solution

```python
import random

def guess(computer_number: int, upper_limit: int)-> None:
  guess: int = -1
  attempt_count: int = 0

  while guess != computer_number:
    guess = int(input(f"Guess number in between range(inclusive) 0 - {upper_limit}: "))
    attempt_count += 1
    if guess == computer_number:
      print(f"Congrats..You guessed the correct number {computer_number} in {attempt_count} attempt")
      break
    elif guess < computer_number:
      print("Sorry...your guess is LOW!")
    elif guess > computer_number:
      print("Sorry...your guess is HIGH!")
    print(f"Attempt: {attempt_count}", end='\n\n')


def main():
  print("=====Guess The Number of Computer=====", end='\n')
  max: int = int(input("Enter the upper limit of the range: "))
  computer_number: int = random.randint(0, max)

  guess(computer_number, max)



if __name__ == "__main__":
  main()
```
