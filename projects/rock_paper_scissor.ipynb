## Rock, Paper Scissor

Rock, Paper, Scissors is a simple hand game played between two players. Each player simultaneously chooses one of three options: Rock, Paper, or Scissors. The winner is determined by the following rules:

- Rock beats Scissors
- Scissors beats Paper
- Paper beats Rock

If both players choose the same option, the game results in a tie. The objective of the game is to determine the winner based on the playersâ€™ choices according to these rules.

## Solution

```python
import random
choice: list = ['r','s','p']

def is_win(player: str, opponent: str) -> bool:
  if (player == 'r' and opponent == 's') or (player == 's' and opponent == 'p') or (player == 'p' and opponent == 'r'):
    return True
  else:
    return False

def play()-> None:
  computer_choice: str = random.choice(choice)
  print("The computer has made its choice.")
  print("Now your turn..", end="\n\n")

  while True:
    user_choice: str = input("Enter Choice Rock(r) | Paper (p) | Scissor(s): ").lower()
    if user_choice in choice:
      break
    else:
      print("Invalid Input", end="\n\n")

  if computer_choice == user_choice:
    print("Both choices are same, hence Tie")
  else:
    user_status: bool = is_win(user_choice, computer_choice)  # Here selecting winner through slightly different appraoch
    print("User status is: ", user_status)

  # selecting winner or looser through different approach
  if computer_choice == 'r' and user_choice == 's':
    print("Computer choice is Rock, hence computer wins")
  elif user_choice == 'r' and computer_choice == 's':
    print("Computer choice is Scissor, hence You wins")

  if computer_choice == 's' and user_choice == 'p':
    print("Computer choice is Scissor, hence computer wins")
  elif user_choice == 's' and computer_choice == 'p':
    print("Computer choice is Paper, hence You wins")

  if computer_choice == 'p' and user_choice == 'r':
    print("Computer choice is Paper, hence computer wins")
  elif user_choice == 'p' and computer_choice == 'r':
    print("Computer choice is Rock, hence You wins")

def main():
  print("=====ROCK, PAPER AND SCISSOR=====", end='\n\n')
  play()

if __name__ == "__main__":
  main()
```
