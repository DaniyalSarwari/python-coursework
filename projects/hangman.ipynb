## Hangman

Hangman is a word-guessing game where the computer randomly selects a secret word, and the player tries to guess it one letter at a time. The player has a limited number of incorrect guesses. Each wrong guess brings the player closer to losing the game. The objective is to correctly guess all the letters in the word before running out of attempts.

## Solution

```python
import requests
from requests import Response
import random

api_url: str = "https://random-word-api.herokuapp.com/word?number=42"

def get_random_word() -> str:
  words: list[str] = ["apple", "cycle", "computer", "python"]
  # word: str
  response: Response = requests.get(api_url)
  if response.status_code == 200:
    # print(response.json())
    # print(type(response.json()))
    words = response.json()
    words = [word for word in words if word.isalpha()]
  else:
    print("Note: Failed to fetch a random word from the API, so a word is selected from a hardcoded list instead.")

  word: str = random.choice(words)
  return word

def get_input() -> str:
  while True:
    guess: str = input("Guess a letter: ")
    if guess.isalpha() == True:
      break
    else:
      print("Please enter a valid characters (A-Z)")

  return guess.upper()


def main():
  word: str = get_random_word().upper()
  # print(word)
  current_status: list = list(len(word) * '-')
  print(' '.join(current_status), end='\n\n')
  lives: int = 5
  used_char: set = set()



  while lives != 0:
    print("LIVES: ", lives)
    print("Character Used: ", ''.join(used_char))

    guess: str = get_input()

    if guess in used_char:
      print("You already utilized this character\n\n")
      continue
    else:
      used_char.add(guess)

    if guess in word:
      word_lst: list = list(word)
      for counter in range(0, len(word_lst)):
        if word_lst[counter] == guess:
          current_status[counter] = guess
    else:
      lives -= 1
    print("Current Word: ", ' '.join(current_status), end='\n\n')
    if '-' not in current_status:
        break

  if (lives == 0) or ('-' in current_status):
    print("\n\n---You Lost---")
    print("The Actual word is: ", word)
  else:
    print("\n\n---You Did It!---")


if __name__ == "__main__":
  main()
```
