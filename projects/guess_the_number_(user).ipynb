## Guess the Number (User)

- The user randomly selects a number within a given range.
- The computer guesses the number, and the user provides feedback if the guess is too high, too low, or correct.
- The game continues until the correct number is guessed, and the total number of attempts is displayed.

## Solution

```python
import random

def computer_guess(upper_limit: int)-> None:
  low: int = 0
  high: int = upper_limit
  feedback: str = ''
  attempt_count: int = 0

  while feedback != 'c':
    print(f"RANGE NOW: ({low}, {high})")
    guess: int = random.randint(low, high)
    feedback = input(f"Is {guess} number is Correct(c), Hight(h) or Low(l) : ").lower()
    attempt_count += 1

    if feedback == 'c':
      print(f"Yay!..I (computer) guessed the correct number {guess} in {attempt_count} attempt")
      break
    elif feedback == 'l':
      print("Ahh!...so your number is too HIGH!")
      low = guess
    elif feedback == 'h':
      print("Ahh!...so your number is too LOW!")
      high = guess
    print(f"Attempt: {attempt_count}", end='\n\n')


def main():
  print("=====Your number guessed by Computer=====", end='\n\n')
  max: int = int(input("Enter the upper limit of the range: "))

  computer_guess(max)

if __name__ == "__main__":
  main()
```
