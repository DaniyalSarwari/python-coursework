## Problem Statement

- Manually creating fun, dynamic text content is repetitive and inefficient
- Users need an easy way to generate interactive Mad Libâ€“style descriptions
- The system should allow placeholders for nouns, verbs, and adjectives
- The solution should be simple, lightweight, and easy to extend

## Solution

```python
def madlibs():
  noun1: str = input("Enter Noun: ")
  noun2: str = input("Enter Noun: ")
  verb1: str = input("Enter Verb: ")
  verb2: str = input("Enter Verb: ")
  adjective1: str = input("Enter Adjective: ")
  adjective2: str = input("Enter Adjective: ")

  story: str = f"""
  Today I woke up feeling incredibly {adjective1}.
  I grabbed my favorite {noun1} and decided it was finally time to {verb1}.
  Everyone around me stared in shock as I continued to {verb2} with pure confidence.
  By the end of the day, my {adjective2} adventure became the most unforgettable {noun2} anyone had ever seen.
  """

  print(story)

if __name__=="__main__":
  madlibs()
```
