```python
from typing import Any
import qrcode
from pyzbar.pyzbar import decode
from PIL import Image

def generate_qr_code(data: Any, filename: str) -> None:
  qr = qrcode.QRCode(
      version=None,
      box_size=5,
      border=2,
      error_correction=qrcode.constants.ERROR_CORRECT_L
  )
  qr.add_data(data=data)
  qr.make(fit=True)
  img = qr.make_image(fill_color="black", back_color="yellow")
  img.save(filename)

def decode_qr_code(filename: str)-> Any:
  try:
    img = Image.open(filename)
    decoded_data = decode(img)

    return decoded_data[0].data.decode("utf-8")
  except:
    return f"Failed to decode QR Code, Filename: {filename}"

def main():
  data: str = "My python learning journey projects!"
  generate_qr_code(data, "qrcode.png")

  result = decode_qr_code("qrcode.png")
  print(result)

if __name__ == "__main__":
  main()
```
