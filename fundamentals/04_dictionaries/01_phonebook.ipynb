## Problem Statement

In this program we show an example of using dictionaries to keep track of information in a phonebook.
This program saves the person's name and number in a phonebook and then read the number against name.

## Solution

```python
# This program is intended to practice dictionary so edge cases/exceptions not handled


def read_phone_book() -> dict:
  phonebook: dict[str,int] = {}
  person_name: str
  person_number: int
  while True:
    user_input: str = input("Enter Person Name: ")
    if user_input == "":
      break
    else:
        person_name = user_input.lower()
        person_number = int(input("Enter Person Number: "))

    phonebook[person_name] = person_number

  return phonebook

def print_all_phonebook(pb: dict) -> None:
  for name, number in pb.items():
    print(f"{name} -> {number}")

def lookup_number(pb: dict) -> None:
  user_name: str = input("Enter Name to Lookup Number: ").lower()
  if user_name not in pb:
    print("No record found!")
  else:
    print(f"{user_name} -> {pb[user_name]}")

def main():
  phonebook = read_phone_book()  # no need to return dictionary since same dictionary is used
  print_all_phonebook(phonebook)
  lookup_number(phonebook)

if __name__ == '__main__':
    main()
```
