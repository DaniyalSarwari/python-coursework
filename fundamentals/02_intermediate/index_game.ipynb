### Objective:
Create a Python program that helps you practice accessing and manipulating elements in a list. This exercise will help you get comfortable with indexing, slicing, and modifying list elements.

#### Initialize a List:
Create a list with at least 5 different elements. They can be numbers, strings, or a mix of both.

#### Accessing Elements:
Write a function that:

#### Modifying Elements:
Write a function that:

- Accepts a list, an index, and a new value as inputs.

#### Slicing the List:
Write a function that:

- Accepts a list, a start index, and an end index as inputs.

- Handles cases where the indices are out of range.

#### Game Interaction:
Create a simple text-based game that:

- Prompts the user to select an operation (access, modify, slice).

## Solution

````python
# Problem 1 Solution
def main():
  fruit_list: list = ['apple', 'banana', 'orange', 'grape', 'pineapple']
  print(fruit_list)
  print("Length: ", len(fruit_list))
  fruit_list.append('mango')
  print(fruit_list)

if __name__ == "__main__":
  main()
````

````python
# Problem 2 Solution
from typing import Any

def get_item(lst: list, index: int)-> Any:
  try:
    return lst[index]
  except IndexError:
    return "Wrong Index"

def update_item(lst: list, index: int, value: Any)-> str | None:
  try:
    lst[index] = value
  except IndexError:
    return "Wrong Index"

def get_slice(lst: list, start: int, end: int)-> list:
  return lst[start:end]

def main():
  items: list = ['belt', 5, [3.14, 'PI'], {'Make':'Hnnda'}, ('red', 'green', 'blue')]
  print("Origional List: ", items)

  print("===Game Interaction===")
  selection: str = input("Select an operation (access, modify, slice): ")
  start_index: int = int(input("Enter Start Index: "))

  if selection.lower() == 'access':
    item: Any = get_item(items, start_index)
    if item != 'Wrong Index':
      print("Value at Index 3: ", item)
    else:
      print(item)

  if selection.lower() == 'modify':
    new_value: str = input("Enter New Value: ")
    item: str | None = update_item(items, start_index, new_value)
    if item == None:
      print("Updated List: ", items)
    else:
      print(item)

  if selection.lower() == 'slice':
    end_index: int = int(input("Enter End Index: "))
    sublist: list = get_slice(items, start_index, end_index)
    print(sublist)

if __name__ == "__main__":
  main()
````
