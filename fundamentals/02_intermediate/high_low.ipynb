## Solution

````python
import random

NUM_ROUNDS: int = 6
score: int = 0

def main():
  global score
  print("Welcome to the High-Low Game!")
  print('-'*30)
  for count in range(1, NUM_ROUNDS+1):
    print("Round ",count)
    computer_number: int = random.randint(1,100)
    user_number: int = random.randint(1,100)
    # print("The computer's number is ",computer_number)
    print("Your number is ",user_number)
    result: str = 'higher' if user_number > computer_number else 'lower'
    choice: str = input("Do you think your number is higher or lower than the computer's?: ")
    choice = choice.lower()

    while True:
      options: list = ['higher', 'lower']
      if choice not in options:
        choice = input("Please enter either higher or lower: ")
      else:
        break

    if choice.lower() == result:
      print("You were right! The computer's number was ", computer_number)
      score += 1
    else:
      print("Aww, that's incorrect. The computer's number was ", computer_number)
    print("Your score is now ", score)
    print(end='\n')

  half_of_rounds: int = round(NUM_ROUNDS /2, 0)
  # print(half_of_rounds)
  half_of_scores: int = int(round(score /2,0))
  # print(half_of_scores)

  if half_of_scores >= half_of_rounds:
    print("Wow! You played perfectly!")
  elif (half_of_scores < half_of_rounds) and (half_of_scores > 1 ):
    print("Good job, you played really well!")
  else:
    print("Better luck next time!")




if __name__ == "__main__":
  main()
````
