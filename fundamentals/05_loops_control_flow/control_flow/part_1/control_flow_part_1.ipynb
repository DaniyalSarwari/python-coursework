```python
# Treasure Island Adventure Game
# Text-based adventure games use decision trees to create branching storylines.
# Each choice leads to different outcomes. This teaches sequential decision-making and if/else logic.

print("==== WELCOME TO TREASURE ISLAND ====")
print("Your mission is to find the treasure.",end='\n\n')
print("-" * 40)
print("You are on a beach with a strange island ahead. You see a path splitting into two directions.")
print("-" * 40,end='\n\n')

decision = input("Do you go left or right? ")
decision = decision.lower()

if decision != "left":
  print("You fall into a hole. Game Over.")
else:
  print("You go left. The path opens to a river.",end="\n\n")

decision = input("Do you swim or wait? ")
decision = decision.lower()

if decision != "wait":
  print("You are attacked by a trout. Game Over.")
else:
  print("You wait by the river. Suddenly, you hear mysterious sounds...", end="\n\n")
  print("A cave reveals three doors: one glowing red, one deep blue, one bright yellow.")
  decision = input("Which door do you choose? (red/blue/yellow): ")
  decision = decision.lower()

  match decision:
    case "red":
      print("\nYou are burned by fire. Game Over.")
    case "blue":
      print("\nYou are eaten by beasts. Game Over.")
    case "yellow":
      print("\n⭐ CONGRATULATIONS! ⭐\n")
      print("You found the treasure and escaped the island!")
      print("=== GAME OVER ===")
    case _:
      print("\nInvalid choice. Game Over.")
```
