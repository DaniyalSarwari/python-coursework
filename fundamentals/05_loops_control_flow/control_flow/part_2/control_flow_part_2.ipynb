```python
# Rollercoaster Pricing System - 4 Stages

print("=== ROLLERCOASTER BOOKING SYSTEM ===",end="\n")
print("Welcome to the ultimate theme park experience!\n")
MAX_HEIGHT:int = 120
PHOTOS_PRICE: int = 3
PRICE_LIST: dict[str, int] = { "underage" : 5, "teenager": 7, "adult" : 12, "senior" : 0}

metrics: dict[str, int] = {"customer_processed": 0, "revenue": 0, "photo_packages": 0, "senior_rode": 0}

while True:
  customer_number = metrics["customer_processed"] + 1
  print(f"\nCustomer {customer_number}")
  print('_'*10)

  photos_check: bool = False
  senior_check: bool = False
  total_price: int = 0
  ticket_price: int
  buy_photos: str
  user_height: int

  while True:
    try:
      user_height: int = int(input("What is your height (in cm)? "))
      break
    except ValueError:
      print("Please Enter a valid numbers!")


  if user_height <= MAX_HEIGHT:
    print("""âœ— Sorry, you are too short to ride.
    Minimum height required: 120cm""")
  else:
    print("âœ“ Great! You can ride this rollercoaster!",end="\n\n")
    age: int

    while True:
      try:
        age = int(input("How old are you? "))
        break
      except ValueError:
        print("Please Enter a valid input number!")

    if age < 12:
      ticket_price:int = PRICE_LIST["underage"]
      print("Age group: Child (Under 12)","\n",f"Ticket price: ${ticket_price}")
    elif 12 <= age <= 18:
      ticket_price:int = PRICE_LIST["teenager"]
      print("Age group: Teenager (12-18)","\n",f"Ticket price: ${ticket_price}")
    elif 18 < age < 45:
      ticket_price:int = PRICE_LIST["adult"]
      print("Age group: Adult (18+)","\n",f"Ticket price: ${ticket_price}")
    elif 45 <= age <= 55:
      senior_check = True
      metrics["senior_rode"] += 1
      ticket_price:int = PRICE_LIST["senior"]
      print("Age group: Senior (45-55)","\n","ðŸŽ‰ SPECIAL DISCOUNT! ðŸŽ‰","\n","Your ride is completely FREE!")

    total_price += ticket_price

    while True:
      try:
        buy_photos = input("\nDo you want to buy photos? (yes/no) ")
        assert buy_photos.lower() in ("yes", "no")
        if buy_photos.lower() == "yes":
          photos_check = True
          metrics["photo_packages"] += 1
          total_price += PHOTOS_PRICE

        break
      except:
        print("Please choose option between (yes/no)!")

    print("\n==== YOUR BILL ====")
    print(f"Ticket: {"$0 (Senior Discount!)" if senior_check else "$"+str(ticket_price)}")
    print(f"Photos: { "+$" + str(PHOTOS_PRICE) if photos_check else "$0" }")
    print("-"*12)
    print(f"Total: ${total_price}")
    if senior_check:
      print("Thank you for celebrating with us!")

    metrics["revenue"] += total_price

    metrics["customer_processed"] = customer_number

  new_customer:str
  while True:
    try:
      new_customer:str = input("\nProcess another customer? (yes/no) ")
      assert new_customer.lower() in ("yes", "no")
      break
    except:
      print("Please choose option between (yes/no)!")

  if new_customer.lower() == "yes":
    continue;
  elif new_customer.lower() == "no":
    print("\n=== SESSION SUMMARY ===")
    print(f"Total Customers Processed: {metrics["customer_processed"]}")
    print(f"Total Revenue: ${metrics["revenue"]}")
    print(f"Photo Packages Sold: {metrics["photo_packages"]}")
    print(f"Senior Free Rides: {metrics["senior_rode"]}")
    print(f"Average Revenue per Customer: ${metrics["revenue"]/2:.2f}")
    print("Thank you for visiting!")
    break


```
