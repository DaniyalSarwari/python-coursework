## Problem Statement

10 even 11 odd 12 even 13 odd 14 even 15 odd 16 even 17 odd 18 even 19 odd

## Solution

```python
def parity_check(number_1: int)-> int:
  if number_1 % 2 == 0:
    return "even"
  else:
    return "odd"

def main():
  for num in range(20):
    print(f"{num} {parity_check(num)}")

if __name__ == '__main__':
    main()
```
