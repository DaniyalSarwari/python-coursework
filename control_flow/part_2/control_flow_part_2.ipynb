```python
# Rollercoaster Pricing System - 4 Stages

MAX_HEIGHT:int = 120

print("=== ROLLERCOASTER BOOKING ===",end="\n\n")
user_height:int = int(input("What is your height (in cm)?"))

if user_height > MAX_HEIGHT:
  print("✓ Great! You can ride this rollercoaster!")
else:
  print("""✗ Sorry, you are too short to ride.
  Minimum height required: 120cm""")
```
