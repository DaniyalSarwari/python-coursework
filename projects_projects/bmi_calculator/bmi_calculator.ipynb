```python
# This program is to check if someone is underweight or overweight through the BMI calculator formula to calculate BMI:

person_weight = float(input("Enter your weight in kilograms: "))
person_height = float(input("Enter your height in meters: "))

bmi = person_weight / (person_height * person_height)
print(f"Your BMI is: {bmi:.02f}")

if bmi < 18.5:
  print("Person is Underweight")
elif 18.5 <= bmi <= 24.9:
  print("Person is Normal weight")
elif 25 <= bmi <= 29.9:
  print("Person is Overweight")
else:
  print("Obese")
```
